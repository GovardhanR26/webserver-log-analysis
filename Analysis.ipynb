```python
import pandas as pd
import matplotlib.pyplot as plt
```

# 1. Which are the IP addresses that have made the most requests?

```python
# Read the text file
df = pd.read_csv("ip.txt", sep="\t", header=None, names=["IP", "Count"])
```

```python
df
```

```python
# Sort in descending order according to the counts
sorted_df = df.sort_values(by=['Count'], ascending=False)
```

```python
sorted_df
```

<h3> Pie Chart showing <i>top 10 IP addresses</i> that have made requests</h3>

```python
plt.pie(sorted_df['Count'].values.tolist()[:10], labels = sorted_df['IP'].values.tolist()[:10], startangle = 90)
plt.show()
```
